```python
n, A, B = parse_dat_file("kra30a")
assert n == 30
assert A.shape == (n, n)
assert B.shape == (n, n)
p = np.random.permutation(n)
# Reorder rows by p, then reorder columns by p
# So move B_{p(i),p(j)} to B′_{i,j}
B_p = B[p][:, p]
# Then doing A * B′ is doing A_{i,j} * B_{p(i),p(j)}
np.sum(A * B_p)
```
